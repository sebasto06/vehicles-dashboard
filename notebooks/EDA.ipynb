```python
# Importo librerias
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importo dataset
df = pd.read_csv('../vehicles_us.csv')
```

### Limpieza de datos

```python
fillna_and_int = ['is_4wd', 'cylinders', 'model_year', 'odometer']
for i in fillna_and_int:
  df[i] = df[i].fillna(0)
  df[i] = df[i].astype(int)

df["paint_color"] = df["paint_color"].fillna("unknown")
```

```python
def cylinder_check(row):
    '''Esta función verificará si falta el valor de cilindros y, si es así, lo completará con el número de cilindros más común para ese modelo de auto y año'''
    if row["cylinders"] == 0:
        cylinder = df.loc[df["model"] == row["model"]].loc[df["model_year"] == row["model_year"]]["cylinders"].unique()
        for i in range(len(cylinder)):
            if cylinder[i] != 0:
                return cylinder[i]
    return row["cylinders"]

df["cylinders"] = df.apply(cylinder_check, axis=1)
```

```python
def odometer_check(row):
    '''Esta función verificará si falta el valor del odómetro y, si es así, lo completará con el valor de odómetro más común para ese modelo de auto y año'''
    if row["odometer"] == 0:
        odometer = df.loc[df["model"] == row["model"]].loc[df["model_year"] == row["model_year"]].loc[df["type"] == row["type"]]["odometer"].unique()
        for i in range(len(odometer)):
            if odometer[i] != 0:
                return odometer[i]
    return row["odometer"]

df["odometer"] = df.apply(odometer_check, axis=1)
```

```python
def model_year_check(row):
    '''Esta función verificará si falta el año del modelo y, si es así, lo completará con el año de modelo más común para ese modelo de auto y año'''
    if row["model_year"] == 0:
        model_year = df.loc[df["model"] == row["model"]].loc[df["type"] == row["type"]]["model_year"].unique()
        for i in range(len(model_year)):
            if model_year[i] != 0:
                return model_year[i]
    return row["model_year"]

df["model_year"] = df.apply(model_year_check, axis=1)
```

```python
def paint_color_check(row):
    '''Esta función verificará si falta el color de la pintura y, si es así, lo completará con el color de pintura más común para ese modelo de auto.'''
    if row["paint_color"] == "unknown":
        paint_color = df.loc[df["model"] == row["model"]]["paint_color"].unique()
        for i in range(len(paint_color)):
            if paint_color[i] != "unknown":
                return paint_color[i]
    return row["paint_color"]

df["paint_color"] = df.apply(paint_color_check, axis=1)
```

```python
df["manufacturer"] = df["model"].str.split(" ", n = 1, expand = True)[0]
```

```python
df["model_name"] = df["model"].str.split(" ", n = 1, expand = True)[1]
```

```python
df.drop(columns=["model"], inplace=True)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
sns.boxplot(x=df["price"], ax=ax[0])
sns.boxplot(x=df["odometer"], ax=ax[1])
plt.show()
```

##### Teniendo en cuenta los boxplots, se identifican algunos outliers que a continuacion se proceden a eliminar con el método IQR

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR)))]
```

```python
Q1 = df['odometer'].quantile(0.25)
Q3 = df['odometer'].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df['odometer'] < (Q1 - 1.5 * IQR)) | (df['odometer'] > (Q3 + 1.5 * IQR)))]
```

```python
df["date_posted"] = pd.to_datetime(df["date_posted"], format='%Y-%m-%d')
```
